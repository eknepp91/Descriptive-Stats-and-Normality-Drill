```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Exponential Distribution

exponential = np.random.exponential(1, 100)

plt.hist(exponential)
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() - exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Rayleigh Distribution
rayleigh = np.random.rayleigh(3, 100)

plt.hist(rayleigh)
plt.axvline(rayleigh.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rayleigh.mean() - rayleigh.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rayleigh.mean() + rayleigh.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Poisson Distribution 
poisson = np.random.poisson(2, 100)

plt.hist(poisson)
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Lognormal
lognormal = np.random.lognormal(0.5, 1, 100)

plt.hist(lognormal)
plt.axvline(lognormal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(lognormal.mean() - lognormal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(lognormal.mean() + lognormal.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Standard Cauchy
standard_cauchy = np.random.standard_cauchy(100)

plt.hist(standard_cauchy)
plt.axvline(standard_cauchy.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standard_cauchy.mean() - standard_cauchy.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(standard_cauchy.mean() + standard_cauchy.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Standard T
standard_t = np.random.standard_t(2,100)

plt.hist(standard_t)
plt.axvline(standard_t.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standard_t.mean() - standard_t.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(standard_t.mean() + standard_t.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Normal Distributions
normal_1 = np.random.normal(5, 0.5, 100)
normal_2 = np.random.normal(10, 1, 100)
normal_3 = normal_1+normal_2

plt.hist(normal_3)
plt.axvline(normal_3.mean(), color='g', linestyle='solid', linewidth =2)
plt.axvline(normal_3.mean() + normal_3.std(), color='g', linestyle='dashed', linewidth =2)
plt.axvline(normal_3.mean() - normal_3.std(), color='g', linestyle='dashed', linewidth =2)

plt.show()

#The results are slightly skewed
```
